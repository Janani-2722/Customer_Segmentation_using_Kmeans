```python
num_cols = ['Promo2SinceWeek', 'Promo2SinceYear', 'CompetitionOpenSinceYear', 'CompetitionOpenSinceMonth']
cat_cols = ['PromoInterval']

store_data[num_cols] = store_data[num_cols].fillna(0)
store_data[cat_cols] = store_data[cat_cols].fillna("None")

```

```python
store_data.isnull().sum()
```

```python
store_data['CompetitionDistance']= store_data['CompetitionDistance'].fillna(store_data['CompetitionDistance'].mean(), inplace = True)
```

```python
store_data.isnull().sum()
```

```python
store_data.hist(bins=30, figsize = (20,20), color='red', edgecolor='black')
plt.show()
```

```python
train_store_data = pd.merge(train_data, store_data, how = 'inner', on = 'Store')
```

```python
train_store_data.to_csv('merged_train_store.csv', index=False)
```

```python
train_store_data.info()
```

```python
correlations = train_store_data.corr(numeric_only=True)['Sales'].sort_values()
print (correlations)
```

```python
corr_matrix = train_store_data.corr(numeric_only=True)
plt.figure(figsize=(20,20))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
train_store_data['Date'] = pd.to_datetime(train_store_data['Date'])
train_store_data.dtypes
```

```python
train_store_data['Year'] = train_store_data['Date'].dt.year
train_store_data['Month'] = train_store_data['Date'].dt.month
train_store_data['Day'] = train_store_data['Date'].dt.day
```

```python
train_store_data.head()
```

```python
avg_sales = train_store_data.groupby('Month')[['Sales']].mean()
avg_sales
```

```python
avg_sales.plot(figsize = (15,10), marker = 'o', color = 'blue')
plt.title = ('Average Sales Per Month')
plt.show()
```

```python
avg_cust = train_store_data.groupby('Month')[['Customers']].mean()
avg_cust
```

```python
avg_cust.plot(figsize = (15,10), marker = '^', color = 'green')
plt.title = ('Average Customers Per Month')
plt.show()
```

```python
sales_per_day = train_store_data.groupby('Day')[['Sales']].mean()
sales_per_day .plot(figsize = (20,5), marker = 'o', color = 'red')
plt.title = ('Average Sales Per Day')
plt.show()
```

```python
cust_per_day = train_store_data.groupby('Day')[['Customers']].mean()
cust_per_day .plot(figsize = (20,5), marker = '^', color = 'orange')
plt.title = ('Average Customers Per Day')
plt.show()
```

```python
sales_per_week = train_store_data.groupby('DayOfWeek')[['Sales']].mean()
sales_per_week .plot(figsize = (20,5), marker = 'o', color = 'green')
plt.title = ('Average Sales Per Day of the Week')
plt.show()
```

```python
cust_per_week = train_store_data.groupby('DayOfWeek')[['Customers']].mean()
cust_per_week .plot(figsize = (20,5), marker = '^', color = 'blue')
plt.title = ('Average Customers Per Day of the Week')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
train_store_data.groupby(['Date', 'StoreType'])['Sales'].mean().unstack().plot(ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
train_store_data.groupby(['Date', 'StoreType'])['Sales'].mean().plot(ax=ax)
plt.show()
```
